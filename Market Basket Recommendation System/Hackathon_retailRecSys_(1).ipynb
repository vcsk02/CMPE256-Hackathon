```python
"""
Market Basket Analysis Recommendation System
San José State University - Data Science Project

This system provides real-time product recommendations based on:
- Association Rule Mining (Apriori & FP-Growth)
- Confidence and Lift metrics
- Dynamic cart analysis

Author: SJSU Data Science Student
Date: 2025
"""

# ============================================================================
# PART 1: INSTALLATION AND IMPORTS
# ============================================================================

# Run this first in Google Colab or Jupyter Notebook
# !pip install mlxtend fastapi uvicorn pandas numpy matplotlib seaborn plotly

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import plotly.express as px
import plotly.graph_objects as go
from plotly.subplots import make_subplots
import warnings
warnings.filterwarnings('ignore')

# Association Rule Mining
from mlxtend.frequent_patterns import apriori, association_rules, fpgrowth
from mlxtend.preprocessing import TransactionEncoder

# API Development
from fastapi import FastAPI, HTTPException
from pydantic import BaseModel
from typing import List, Optional, Dict
import uvicorn

# Utilities
from collections import Counter, defaultdict
from datetime import datetime
import pickle
import json


# -------- Co-occurrence helper (fallback when rules don't match) --------
from collections import Counter, defaultdict
from itertools import combinations

def build_cooccurrence(transactions):
    """
    Build item frequency and pair co-occurrence counts from transactions.
    Returns: item_counts (Counter), pair_counts (Counter of (a,b) with a<b)
    """
    item_counts = Counter()
    pair_counts = Counter()

    for t in transactions:
        u = sorted(set(t))  # unique items per basket
        item_counts.update(u)
        for a, b in combinations(u, 2):
            pair_counts[(a, b)] += 1

    return item_counts, pair_counts

# ============================================================================
# PART 2: DATA LOADING AND PREPROCESSING
# ============================================================================

class MarketBasketDataProcessor:
    """
    Handles data loading, cleaning, and transformation
    """

    def __init__(self, file_path: str):
        self.file_path = file_path
        self.df = None
        self.transactions = None
        self.products = None

    def load_data(self):
        """Load the CSV file"""
        print("Loading data...")
        self.df = pd.read_csv(self.file_path)
        print(f"Loaded {len(self.df)} transactions")
        return self.df

    def explore_data(self):
        """Basic data exploration"""
        print("\n" + "="*60)
        print("DATA EXPLORATION")
        print("="*60)
        print(f"\nDataset Shape: {self.df.shape}")
        print(f"\nFirst few rows:")
        print(self.df.head())
        print(f"\nData Info:")
        print(self.df.info())
        print(f"\nMissing Values:")
        print(self.df.isnull().sum())

    def extract_product_names(self, item_string):
        """Extract product name from 'Product Name (SKU: XXX)' format"""
        if pd.isna(item_string):
            return None
        # Extract everything before ' (SKU:'
        return item_string.split(' (SKU:')[0].strip()

    def preprocess_data(self):
        """
        Convert data into transaction format
        Each row = one transaction
        Each transaction = list of products
        """
        print("\nPreprocessing data...")

        # Get all item columns
        item_columns = [col for col in self.df.columns if col.startswith('item_')]

        # Create transaction list
        transactions = []
        for idx, row in self.df.iterrows():
            transaction = []
            for col in item_columns:
                if pd.notna(row[col]):
                    # Extract product name only
                    product = self.extract_product_names(row[col])
                    if product:
                        transaction.append(product)
            if transaction:  # Only add non-empty transactions
                transactions.append(transaction)

        self.transactions = transactions

        # Get unique products
        self.products = set()
        for transaction in transactions:
            self.products.update(transaction)

        print(f"Total transactions: {len(self.transactions)}")
        print(f"Unique products: {len(self.products)}")

        return transactions

    def create_basket_dataframe(self):
        """
        Create one-hot encoded dataframe for Apriori algorithm
        """
        print("\nCreating basket dataframe...")

        # Use TransactionEncoder from mlxtend
        te = TransactionEncoder()
        te_ary = te.fit(self.transactions).transform(self.transactions)
        basket_df = pd.DataFrame(te_ary, columns=te.columns_)

        print(f"Basket dataframe shape: {basket_df.shape}")
        return basket_df


# ============================================================================
# PART 3: MARKET BASKET ANALYSIS ENGINE
# ============================================================================

class MarketBasketAnalyzer:
    """
    Performs association rule mining and generates recommendations
    """

    def __init__(self, basket_df, min_support=0.01, min_confidence=0.3):
        self.basket_df = basket_df
        self.min_support = min_support
        self.min_confidence = min_confidence
        self.frequent_itemsets = None
        self.rules = None

    def find_frequent_itemsets(self, algorithm='apriori'):
        """
        Find frequent itemsets using Apriori or FP-Growth
        """
        print(f"\nFinding frequent itemsets using {algorithm}...")

        if self.basket_df is None or self.basket_df.empty:
            print('[ERROR] Basket DataFrame is empty.'); return None

        if algorithm == 'apriori':
            self.frequent_itemsets = apriori(
                self.basket_df,
                min_support=self.min_support,
                use_colnames=True
            )
        elif algorithm == 'fpgrowth':
            self.frequent_itemsets = fpgrowth(
                self.basket_df,
                min_support=self.min_support,
                use_colnames=True
            )

        print(f"Found {len(self.frequent_itemsets)} frequent itemsets")
        if self.frequent_itemsets is None or getattr(self.frequent_itemsets, 'empty', False):
            print('[WARN] Frequent itemsets are empty. Consider lowering min_support or checking data.')
        return self.frequent_itemsets

    def generate_rules(self, metric='confidence', min_threshold=None):
        """
        Generate association rules from frequent itemsets
        """
        if min_threshold is None:
            min_threshold = self.min_confidence

        print(f"\nGenerating association rules...")

        # Guard: no rules if no frequent itemsets
        if self.frequent_itemsets is None or getattr(self.frequent_itemsets, 'empty', False):
            print('[WARN] No frequent itemsets available; cannot generate rules.')
            import pandas as _pd
            self.rules = _pd.DataFrame(columns=['antecedents','consequents','support','confidence','lift'])
            return self.rules

        self.rules = association_rules(
            self.frequent_itemsets,
            metric=metric,
            min_threshold=min_threshold
        )

        # Add lift ratio for better filtering
        self.rules['lift_ratio'] = self.rules['lift']

        print(f"Generated {len(self.rules)} rules")
        return self.rules

    def analyze_rules(self):
        """
        Analyze and display key statistics about the rules
        """
        if self.rules is None or len(self.rules) == 0:
            print("No rules to analyze. Generate rules first.")
            return

        print("\n" + "="*60)
        print("ASSOCIATION RULES ANALYSIS")
        print("="*60)

        print(f"\nTotal Rules: {len(self.rules)}")
        print(f"\nConfidence Statistics:")
        print(self.rules['confidence'].describe())
        print(f"\nLift Statistics:")
        print(self.rules['lift'].describe())
        print(f"\nSupport Statistics:")
        print(self.rules['support'].describe())

        # Top rules by confidence
        print("\n" + "-"*60)
        print("TOP 10 RULES BY CONFIDENCE")
        print("-"*60)
        top_rules = self.rules.nlargest(10, 'confidence')[
            ['antecedents', 'consequents', 'confidence', 'lift', 'support']
        ]
        for idx, row in top_rules.iterrows():
            ant = ', '.join(list(row['antecedents']))
            cons = ', '.join(list(row['consequents']))
            print(f"\nIF: {ant}")
            print(f"THEN: {cons}")
            print(f"Confidence: {row['confidence']:.3f} | Lift: {row['lift']:.3f} | Support: {row['support']:.3f}")


# ============================================================================
# PART 4: RECOMMENDATION ENGINE
# ============================================================================

class RecommendationEngine:
    """
    Real-time recommendation engine for cart-based suggestions
    """



    def __init__(self, rules_df, item_counts=None, pair_counts=None, name_cleaner=None):
      self.rules = rules_df
      self.recommendation_cache = {}
      # fallback data (optional)
      self.item_counts = item_counts
      self.pair_counts = pair_counts
      # cleaning function to normalize names identically to your processor
      self._clean = name_cleaner if name_cleaner is not None else (lambda x: str(x).strip().lower())


    def get_recommendations(
    self,
    cart_items: List[str],
    top_n: int = 5,
    min_confidence: float = 0.3,
    min_lift: float = 1.0
    ) -> List[Dict]:

        if not cart_items:
            return []

        # ---------- cleaner identical to your processor (strip SKU, trim, collapse, lower) ----------
        def _clean_name(x: str) -> str:
            if not isinstance(x, str):
                return ""
            x = x.strip()
            if " (SKU:" in x:
                x = x.split(" (SKU:")[0]
            x = " ".join(x.split()).lower()
            return x

        cart_set_lower = {_clean_name(c) for c in cart_items if isinstance(c, str) and c.strip()}

        # ---------- RULE-BASED ----------
        rule_recs = []
        if self.rules is not None and not getattr(self.rules, "empty", False):
            from collections import defaultdict
            agg = defaultdict(lambda: {'confidence': 0.0, 'lift': 0.0, 'support': 0.0, 'count': 0})

            for _, rule in self.rules.iterrows():
                ants_lower = {_clean_name(s) for s in rule['antecedents']}
                cons_lower = {_clean_name(s) for s in rule['consequents']}
                if ants_lower.issubset(cart_set_lower):
                    new_items_lower = cons_lower - cart_set_lower
                    if new_items_lower and rule['confidence'] >= min_confidence and rule['lift'] >= min_lift:
                        for item_lower in new_items_lower:
                            item_original = next((orig for orig in rule['consequents']
                                                  if _clean_name(orig) == item_lower), item_lower)
                            agg[item_original]['confidence'] = max(agg[item_original]['confidence'],
                                                                  float(rule['confidence']))
                            agg[item_original]['lift'] = max(agg[item_original]['lift'], float(rule['lift']))
                            agg[item_original]['support'] += float(rule['support'])
                            agg[item_original]['count'] += 1

            if agg:
                for item, s in agg.items():
                    score = s['confidence'] * 0.5 + s['lift'] * 0.3 + s['support'] * 0.2
                    rule_recs.append({
                        'product': item,
                        'confidence': round(s['confidence'], 3),
                        'lift': round(s['lift'], 3),
                        'support': round(s['support'], 4),
                        'score': round(score, 3),
                        'frequency': s['count']
                    })
                rule_recs.sort(key=lambda x: x['score'], reverse=True)
                return rule_recs[:top_n]

        # ---------- FALLBACK: CO-OCCURRENCE ----------
        # Requires item_counts & pair_counts built from your transactions
        if self.item_counts and self.pair_counts:
            from collections import defaultdict
            scores = defaultdict(lambda: {'pair': 0, 'den': 0})

            # Normalize cart to original-cased names that exist in item_counts keys
            # We’ll match by cleaned name but keep display pleasant
            # Build a mapping clean->original for items seen in cooccurrence
            name_map = {}
            for it in self.item_counts.keys():
                name_map[_clean_name(it)] = it

            cart_clean = [c for c in cart_set_lower if c in name_map]  # keep only items we know
            if not cart_clean:
                return []

            # For each cart item, score candidates that co-occur with it:
            for c in cart_clean:
                c_orig = name_map[c]
                c_count = self.item_counts.get(c_orig, 0)
                if c_count == 0:
                    continue
                for (a, b), pcount in self.pair_counts.items():
                    # pairs are stored as (a,b) with a<b
                    if a == c_orig or b == c_orig:
                        other = b if a == c_orig else a
                        o_clean = _clean_name(other)
                        if o_clean in cart_set_lower:
                            continue  # already in cart
                        # simple lift-like score: P(other | c) = cooccur / count(c)
                        scores[other]['pair'] += pcount
                        scores[other]['den'] += c_count

            if not scores:
                return []

            recs = []
            for item, s in scores.items():
                # conditional probability estimate aggregated across cart items
                conf_like = s['pair'] / s['den'] if s['den'] else 0.0
                # popularity prior
                sup_like = self.item_counts.get(item, 0) / max(1, sum(self.item_counts.values()))
                # combined score
                combined = 0.7 * conf_like + 0.3 * sup_like
                recs.append({
                    'product': item,
                    'confidence': round(conf_like, 3),
                    'lift': None,
                    'support': round(sup_like, 4),
                    'score': round(combined, 3),
                    'frequency': self.item_counts.get(item, 0)
                })
            recs.sort(key=lambda x: x['score'], reverse=True)
            return recs[:top_n]

        # Nothing to recommend
        return []





    def explain_recommendation(self, cart_items: List[str], recommended_item: str):
        """
        Provide explanation for why an item was recommended
        """
        cart_set = set(cart_items)
        explanations = []

        for idx, rule in self.rules.iterrows():
            antecedents = set(rule['antecedents'])
            consequents = set(rule['consequents'])

            if antecedents.issubset(cart_set) and recommended_item in consequents:
                explanations.append({
                    'based_on': list(antecedents),
                    'confidence': round(rule['confidence'], 3),
                    'lift': round(rule['lift'], 3),
                    'support': round(rule['support'], 4)
                })

        return explanations


# ============================================================================
# PART 5: VISUALIZATION
# ============================================================================

class MarketBasketVisualizer:
    """
    Create visualizations for market basket analysis
    """

    def __init__(self, analyzer, processor):
        self.analyzer = analyzer
        self.processor = processor

    def plot_support_distribution(self):
        """Plot distribution of support values"""
        plt.figure(figsize=(10, 6))
        plt.hist(self.analyzer.frequent_itemsets['support'], bins=50, edgecolor='black')
        plt.xlabel('Support')
        plt.ylabel('Frequency')
        plt.title('Distribution of Support Values')
        plt.grid(True, alpha=0.3)
        plt.show()

    def plot_confidence_lift(self):
        """Plot confidence vs lift scatter"""
        # Guard: rules must exist and be non-empty
        if getattr(self, 'analyzer', None) is None or getattr(self.analyzer, 'rules', None) is None:
            print('[INFO] Skipping plot_confidence_lift: no rules object.')
            return
        if len(self.analyzer.rules) == 0:
            print('[INFO] Skipping plot_confidence_lift: rules DataFrame is empty.')
            return
        if self.analyzer.rules[['confidence','lift','support']].dropna().empty:
            print('[INFO] Skipping plot_confidence_lift: metrics are NaN or empty.')
            return
        plt.figure(figsize=(12, 6))
        plt.scatter(
            self.analyzer.rules['confidence'],
            self.analyzer.rules['lift'],
            alpha=0.5,
            c=self.analyzer.rules['support'],
            cmap='viridis'
        )
        plt.colorbar(label='Support')
        plt.xlabel('Confidence')
        plt.ylabel('Lift')
        plt.title('Confidence vs Lift (colored by Support)')
        plt.grid(True, alpha=0.3)
        plt.axhline(y=1, color='r', linestyle='--', label='Lift = 1')
        plt.legend()
        plt.show()

    def plot_top_products(self, top_n=15):
        """Plot most frequent products"""
        # Guard: ensure transactions exist
        if getattr(self.processor, 'transactions', None) is None or len(self.processor.transactions) == 0:
            print('[INFO] Skipping plot_top_products: no transactions available.')
            return
        # Count product occurrences
        product_counts = Counter()
        for transaction in self.processor.transactions:
            product_counts.update(transaction)

        # Get top products
        top_products = dict(product_counts.most_common(top_n))
        if len(top_products) == 0:
            print('[INFO] Skipping plot_top_products: product counts are empty.')
            return


        plt.figure(figsize=(12, 8))
        plt.barh(list(top_products.keys()), list(top_products.values()))
        plt.xlabel('Frequency')
        plt.ylabel('Product')
        plt.title(f'Top {top_n} Most Frequent Products')
        plt.tight_layout()
        plt.show()

    def plot_rule_metrics_heatmap(self, top_n=20):
        """Create heatmap of top rules"""
        # Guard: ensure rules exist
        if getattr(self, 'analyzer', None) is None or getattr(self.analyzer, 'rules', None) is None:
            print('[INFO] Skipping heatmap: no rules object.')
            return
        if self.analyzer.rules.empty:
            print('[INFO] Skipping heatmap: rules DataFrame is empty.')
            return
        top_rules = self.analyzer.rules.nlargest(top_n, 'confidence')

        # Create rule labels
        rule_labels = []
        for idx, row in top_rules.iterrows():
            ant = ', '.join(list(row['antecedents']))[:30]
            cons = ', '.join(list(row['consequents']))[:30]
            rule_labels.append(f"{ant} → {cons}")

        # Create data matrix
        data = top_rules[['support', 'confidence', 'lift']].values
        if data.size == 0:
            print('[INFO] Skipping heatmap: no data to plot.')
            return


        plt.figure(figsize=(10, 12))
        sns.heatmap(
            data.T,
            xticklabels=rule_labels,
            yticklabels=['Support', 'Confidence', 'Lift'],
            annot=True,
            fmt='.2f',
            cmap='YlOrRd'
        )
        plt.title(f'Top {top_n} Rules - Metrics Heatmap')
        plt.xticks(rotation=90)
        plt.tight_layout()
        plt.show()


# ============================================================================
# PART 6: FASTAPI DEPLOYMENT
# ============================================================================

class RecommendationAPI:
    """
    FastAPI application for recommendation system
    """

    def __init__(self, recommendation_engine, rules_df):
        self.engine = recommendation_engine
        self.rules = rules_df
        self.app = FastAPI(
            title="Market Basket Recommendation API",
            description="Real-time product recommendations based on cart analysis",
            version="1.0.0"
        )
        self.setup_routes()

    def setup_routes(self):
        """Setup API endpoints"""

        # Request/Response models
        class CartRequest(BaseModel):
            cart_items: List[str]
            top_n: Optional[int] = 5
            min_confidence: Optional[float] = 0.3
            min_lift: Optional[float] = 1.0

        class RecommendationResponse(BaseModel):
            recommendations: List[Dict]
            cart_items: List[str]
            timestamp: str

        @self.app.get("/")
        def root():
            return {
                "message": "Market Basket Recommendation API",
                "endpoints": {
                    "/recommend": "POST - Get recommendations",
                    "/explain": "POST - Explain recommendation",
                    "/stats": "GET - Get system statistics"
                }
            }

        @self.app.post("/recommend", response_model=RecommendationResponse)
        def get_recommendations(request: CartRequest):
            """Get product recommendations based on cart"""
            try:
                recommendations = self.engine.get_recommendations(
                    cart_items=request.cart_items,
                    top_n=request.top_n,
                    min_confidence=request.min_confidence,
                    min_lift=request.min_lift
                )

                return RecommendationResponse(
                    recommendations=recommendations,
                    cart_items=request.cart_items,
                    timestamp=datetime.now().isoformat()
                )
            except Exception as e:
                raise HTTPException(status_code=500, detail=str(e))

        @self.app.get("/stats")
        def get_stats():
            """Get system statistics"""
            return {
                "total_rules": len(self.rules),
                "avg_confidence": float(self.rules['confidence'].mean()),
                "avg_lift": float(self.rules['lift'].mean()),
                "timestamp": datetime.now().isoformat()
            }

    def run(self, host="0.0.0.0", port=8000):
        """Run the API server"""
        uvicorn.run(self.app, host=host, port=port)


# ============================================================================
# PART 7: MAIN EXECUTION PIPELINE
# ============================================================================

def main():
    """
    Main execution pipeline (robust mining + co-occurrence fallback)
    """
    import pickle

    print("="*70)
    print("MARKET BASKET RECOMMENDATION SYSTEM")
    print("San José State University - Data Science Project")
    print("="*70)

    # -----------------------------
    # STEP 1: Load & preprocess
    # -----------------------------
    print("\n[STEP 1] Loading and preprocessing data...")
    processor = MarketBasketDataProcessor('CMPE256_Hackathon_market_basket_analysis_Release.csv')
    df = processor.load_data()
    if df is None or getattr(df, "empty", False):
        raise ValueError("[FATAL] Loaded DataFrame is empty or None. Check CSV path/format.")

    processor.explore_data()

    transactions = processor.preprocess_data()
    if transactions is None or len(transactions) == 0:
        raise ValueError("[FATAL] No transactions after preprocessing. "
                         "Check that item columns start with 'item_' or update the prefix in preprocess_data().")

    basket_df = processor.create_basket_dataframe()
    if basket_df is None or getattr(basket_df, "empty", False) or (hasattr(basket_df, "values") and basket_df.values.sum() == 0):
        raise ValueError("[FATAL] Basket DataFrame is empty or all zeros. Check one-hot encoding/grouping.")

    # Sanity: transaction lengths
    lengths = [len(t) for t in processor.transactions]
    print(f"Transactions: {len(lengths)} | mean length: {sum(lengths)/max(1,len(lengths)):.2f} | "
          f">=2 items: {sum(1 for x in lengths if x>=2)}")

    # -----------------------------
    # STEP 2: Mining with auto-tune
    # -----------------------------
    print("\n[STEP 2] Performing market basket analysis...")
    analyzer = MarketBasketAnalyzer(
        basket_df=basket_df,
        min_support=0.01,     # initial try
        min_confidence=0.30   # initial try
    )

    support_grid = [0.01, 0.005, 0.003, 0.001, 0.0005]
    conf_grid    = [0.30, 0.20, 0.10, 0.05]

    found = False
    rules = None

    for sup in support_grid:
        for conf in conf_grid:
            analyzer.min_support = sup
            analyzer.min_confidence = conf
            print(f"  → Trying support={sup}, confidence={conf}")
            analyzer.find_frequent_itemsets(algorithm='apriori')

            if analyzer.frequent_itemsets is None or analyzer.frequent_itemsets.empty:
                print("    No frequent itemsets. Continuing...")
                continue

            rules = analyzer.generate_rules(metric='confidence', min_threshold=conf)
            if rules is not None and not rules.empty:
                print(f"    ✅ Found {len(rules)} rules at support={sup}, confidence={conf}")
                found = True
                break
            else:
                print("    No rules. Continuing...")
        if found:
            break

    if not found:
        print("[ERROR] Could not generate any rules even with loose thresholds.")
        print("       Consider checking item column names or whether most transactions contain only one item.")
        # We will still proceed with fallback-only recommendations
        rules = None

    if rules is not None and not rules.empty:
        analyzer.analyze_rules()

    # -----------------------------
    # STEP 3: Build co-occurrence fallback
    # -----------------------------
    # Uses your transaction data to suggest items that co-occur with cart items.
    # (Requires you to have added the helper build_cooccurrence(...) as suggested.)
    print("\n[STEP 3] Building co-occurrence fallback...")
    try:
        item_counts, pair_counts = build_cooccurrence(processor.transactions)
        print(f"[FALLBACK] Items: {len(item_counts)} | Pairs: {len(pair_counts)}")
    except NameError:
        # If build_cooccurrence is not defined, fallback will simply be disabled.
        print("[WARN] build_cooccurrence(...) not found. Fallback recommender will be disabled.")
        item_counts, pair_counts = None, None

    # -----------------------------
    # STEP 4: Recommendation engine
    # -----------------------------
    print("\n[STEP 4] Creating recommendation engine...")
    rec_engine = RecommendationEngine(
        rules,
        item_counts=item_counts,
        pair_counts=pair_counts
    )

    # -----------------------------
    # STEP 5: Test recommendations
    # -----------------------------
    print("\n[STEP 5] Testing recommendation system...")
    test_scenarios = [
        ["Bosch B5512 Control Panel"],
        ["Honeywell VISTA-20P Control Panel", "DSC PowerSeries Neo Alarm Kit"],
        ["Hikvision DS-7608NI-I2/8P NVR", "Dahua 8-Channel NVR"],
        # includes SKU + spaces (cleaner in get_recommendations should handle)
        [" Axis P3225-LVE Network Camera (SKU: 0935-001) "]
    ]

    for scenario in test_scenarios:
        print(f"\n{'='*60}")
        print(f"Cart Items: {scenario}")
        print(f"{'='*60}")

        recs = rec_engine.get_recommendations(scenario, top_n=5)
        if recs:
            for i, rec in enumerate(recs, 1):
                print(f"\n{i}. {rec['product']}")
                print(f"   Score: {rec['score']} | Confidence: {rec['confidence']} | Lift: {rec['lift']} | Support: {rec['support']}")
        else:
            print("No recommendations available for this combination.")

    # -----------------------------
    # STEP 6: Visualizations (guarded)
    # -----------------------------
    print("\n[STEP 6] Creating visualizations...")
    visualizer = MarketBasketVisualizer(analyzer, processor)
    visualizer.plot_top_products(top_n=20)
    visualizer.plot_confidence_lift()
    visualizer.plot_rule_metrics_heatmap(top_n=15)

    # -----------------------------
    # STEP 7: Save model (if rules exist)
    # -----------------------------
    print("\n[STEP 7] Saving model...")
    try:
        with open('recommendation_engine.pkl', 'wb') as f:
            pickle.dump({
                'rules': rules,
                'frequent_itemsets': None if analyzer is None else analyzer.frequent_itemsets,
                'products': [] if processor.products is None else list(processor.products),
                'cooccurrence_item_counts': item_counts,
                'cooccurrence_pair_counts': pair_counts
            }, f)
        print("Model saved successfully!")
    except Exception as e:
        print(f"[WARN] Could not save model: {e}")



    return processor, analyzer, rec_engine, visualizer




# ============================================================================
# EXECUTION
# ============================================================================

if __name__ == "__main__":
    # Run the complete pipeline
    processor, analyzer, rec_engine, visualizer = main()

    print("\n" + "="*70)
    print("SYSTEM READY!")
    print("="*70)
    print("\nYou can now:")
    print("1. Test recommendations: rec_engine.get_recommendations(['Product Name'])")
    print("2. Explore rules: analyzer.rules.head()")
    print("3. Create visualizations: visualizer.plot_...()")
    print("4. Deploy API: See API deployment section")
```
